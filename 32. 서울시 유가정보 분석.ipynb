### 서울시 유가 정보 분석

```python
# file name 합치기
from glob import glob

station_files = glob("../Data/oil/*.xls")

# file name 정렬하기
station_files = sorted(station_files)
station_files
```

### 파일 읽어오기

```python
import pandas as pd
```

```python
# 연습
pd.read_excel("../Data/oil/지역_위치별(주유소) (1).xls", header=2).head()
```

```python
# File 내용 전부 가져오기 : 하나의 List로 만들기 

tmp_raw = []

for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)

tmp_raw
```

```python
# List -> DataFrame으로 변환

station_raw = pd.concat(tmp_raw)
station_raw.head()
```

```python
# station_raw의 샘플(데이터) 갯수?
len(station_raw)
station_raw.shape[0]
```

```python
# 구성 확인
station_raw.info()
```

```python
# 필요한 항목만 선택해서 DataFrame 구성
# 상호 -> Oil_store, 주소 -> 주소, 휘발유 -> 가격, 셀프여부 -> 셀프, 상표 -> 상표

stations = pd.DataFrame(
    {
        'Oil_store' : station_raw['상호'],
        '주소' : station_raw['주소'],
        '가격' : station_raw['휘발유'],
        '셀프' : station_raw['셀프여부'],
        '상표' : station_raw['상표']
    }
)
stations.head()
```

```python
# Ex: 주소 컬럼에서 주소의 앞부분의 글자를 선택하기
address = "서울특별시 강동구 천호대로 1246 (둔촌제2동)"
address.split()[0]
```

```python
# 주소 컬럼에서 주소의 앞부분의 글자만 리스트로 넣자
temp = []

for address in stations.loc[:,'주소']:
    temp.append(address.split()[0])

temp
```

### 주소의 첫글자 변경

```python
# Ex : 123 -> abc
str = "123defghijklmn"
str = str.replace("123", "abc")
str
```

```python
# 주소의 서울특별시, 서울시 -> 서울
stations.주소 = stations.주소.str.replace("서울특별시", "서울")
stations.주소 = stations.주소.str.replace("서울시", "서울")

set([address.split()[0] for address in stations['주소']])
```

```python
stations.head()
```

```python
# stations의 주소에서 구(자치구) 정보를 선택하여 구 라는 컬럼으로 추가하기
stations['구'] = [address.split()[1] for address in stations.주소]
stations.head()
```

### 지도 표시

```python
import json 
import folium
```

```python
geo_path = "../Data/skorea_municipalities_geo_simple.json"
geo_str = json.load(open(geo_path, encoding="utf-8"))

# 서울시 지도 
map = folium.Map(
    [37.5502, 126.982],
    zoom_start=11
)

# 구별 경계면과 평균가를 색깔로 표시
folium.Choropleth(
    geo_data=geo_str,
    data=gu_data,
    columns=[gu_data.index, '가격'],
    fill_color='PuRd',
    key_on='feature.id'

).add_to(map)

map
```

### 가장 비싼 주유소 Top10과 가장 싼 주유소 Top10 시각화

```python
# 주유소 위도 경도 데이터
stations_latlng = pd.read_csv("../Data/oil/stations_latlng.csv")
stations_latlng.head()
```

```python
stations.head()
```

```python
# 기존 stations에 index기준으로 위,경도 데이터 병합하기
stations = \
    pd.merge(
        stations,
        stations_latlng,
        left_index=True,
        right_index=True
    )
stations.head()
```

```python
# 가격이 비싼 Top10

oil_price_top10 = \
    stations.sort_values(
                    '가격',
                    ascending=False
    ).head(10)

oil_price_top10
```

```python
# 가격이 싼 Top10

oil_price_bottom10 = \
    stations.sort_values(
                    '가격',
                    ascending=True
    ).head(10)

oil_price_bottom10
```

```python
# 지도 표시

geo_path = "../Data/skorea_municipalities_geo_simple.json"
geo_str = json.load(open(geo_path, encoding="utf-8"))

# 서울시 지도 
map = folium.Map(
    [37.5502, 126.982],
    zoom_start=11
)

# 구별 경계면과 평균가를 색깔로 표시
folium.Choropleth(
    geo_data=geo_str,
    data=gu_data,
    columns=[gu_data.index, '가격'],
    fill_color='PuRd',
    key_on='feature.id'

).add_to(map)

# 비싼 주유소 Top10
for i in oil_price_top10.index:
    popup1 = folium.Popup(
                    oil_price_top10.loc[:,"Oil_store"][i],
                    max_width=200
    )

    folium.Marker(
        [oil_price_top10.loc[:,'lat'][i], oil_price_top10.loc[:,'lng'][i]],
        popup=popup1,
        icon=folium.Icon(color='red')
    ).add_to(map)

# 싼 주유소 Top10
for i in oil_price_bottom10.index:
    popup1 = folium.Popup(
                    oil_price_bottom10.loc[:,"Oil_store"][i],
                    max_width=200
    )

    folium.Marker(
        [oil_price_bottom10.loc[:,'lat'][i], oil_price_bottom10.loc[:,'lng'][i]],
        popup=popup1,
        icon=folium.Icon(color='blue')
    ).add_to(map)


map
```
