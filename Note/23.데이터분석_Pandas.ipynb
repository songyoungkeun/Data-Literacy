### Fare를 Category화 하기

```python
# Fare를 3단계로 구분
# Fare의 기본 통계량
train.Fare.describe()
```

```python
# Fare -> LowFare(<15), MediumFare(15~30), HighFare(>30)
train['LowFare'] = train.Fare < 15
train['MediumFare'] = (train.Fare >= 15) & (train.Fare < 30)
train['HighFare'] = train.Fare >= 30

train.loc[:,['Fare', 'LowFare', 'MediumFare', 'HighFare']].head(10)
```

```python
# FareType이라는 컬럼에
# Fare가 15보다 작은면 Low
# Fare가 15~30이면 Medium
# Fare가 30초과하면 High

train.loc[train.Fare < 15, "FareType"] = "Low"
train.loc[(train.Fare >= 15) & (train.Fare < 30), "FareType"] = "Medium"
train.loc[train.Fare > 30, "FareType"] = "High"

train.loc[:,['Fare', 'FareType']].head(10)
```

### Python의 함수를 이용하여 처리

```python
def check_fare_type(fare):
    return_value = ""
    if fare < 15:
        return_value = "Low"
    elif fare >= 15 and fare < 30:
        return_value = "Medium"
    else:
        return_value = "High"
    return return_value
```

```python
train['FareType2'] = train.Fare.apply(check_fare_type)
train.loc[:,['Fare', 'FareType', 'FareType']].head(10)
```
