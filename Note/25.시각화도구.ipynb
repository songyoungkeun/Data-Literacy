### Matplotlib
: 파이썬의 표준 시각화 도구

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
# NaN 처리 : NaN을 만나면 앞데이터로 채움
df.ffill(inplace=True)
df.head()
```

```python
df.info()
```

#### 서울에서 다른 지역으로 이동한 데이터만 추출

```python
# 전출지별이 서울특별시 이고 전입지별이 서울특별시가 아닌 데이터만 발췌
df_seoul = df[(df.전출지별 == '서울특별시') & (df.전입지별 != '서울특별시')]
df_seoul.head()
```

```python
# df_seoul의 전출지별 컬럼 삭제
del df_seoul['전출지별']
df_seoul.head()
```

```python
# 전입지별 컬럼 이름을 전입지로 변경
df_seoul.rename(
    {'전입지별' : '전입지'},
    axis='columns',
    inplace=True
)

df_seoul.head()
```

```python
# df_seoul의 index를 전입지 컬럼으로 변경
df_seoul.set_index("전입지", inplace=True)
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택 

sr_one = df_seoul.loc['경기도',:]
sr_one.head()
```

```python
plt.plot(
    sr_one.index,
    sr_one.values
)

plt.show()
```

```python
plt.plot(
    sr_one
)

plt.show()
```

### 한 화면에 차트 여러개 사용하기

```python
# Style
plt.style.use("ggplot")

# 도화지 : 한 화면
fig = plt.figure(figsize=(10, 10))

# 그래프 객체
ax1 = fig.add_subplot(2, 1, 1) #2행 1열의 1번째
ax2 = fig.add_subplot(2, 1, 2) #2행 1열의 2번째

# 데이터로 차트 출력
ax1.plot(
    sr_one,
    'o',
    markersize=10
)

ax2.plot(
    sr_one,
    marker="o",
    markersize=10,
    markerfacecolor = 'green',
    color = "olive",
    linewidth=2,
    label='서울->경기'
)

ax2.legend(loc='best')

# y축 범위 지정
ax1.set_ylim(0, 800000)
ax2.set_ylim(0, 800000)

# x축 눈금라벨 75도 회전
ax2.set_xticklabels(
    sr_one.index, # Data 
    rotation=75
)

plt.show()
```
