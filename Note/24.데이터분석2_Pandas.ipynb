```python
import pandas as pd
```

```python
data = {
    'names'  : ['kenny', 'kenny', 'kenny', 'charles', 'charles'],
    'year'   : [   2016,    2017,    2018,      2016,      2017],
    'points' : [    1.5,    1.77,     3.6,       2.4,       2.9] 
}

df = pd.DataFrame(
    data,
    columns=['year', 'names', 'points', 'penalty'],
    index=['one','two','three','four','five']
)

df
```

```python
# penalty에 data 넣기
df.penalty = 0.5
df
```

```python
import numpy as np

df.penalty = np.arange(0.1, 0.5+0.1, 0.1)
df
```

```python
df['zeros'] = list(range(0,4+1,1))
df
```

```python
# Series
val = pd.Series(
    [-1.2, -1.5, -1.7],
    index=['two', 'four', 'five']
)
val
```

```python
df['debt'] = val
df
```

```python
# net_points = points - penalty
df['net_points'] = df.points - df.penalty
df
```

```python
# df에 high_points라는 컬럼을 생성 : net_points가 2초과면 True 아니면 False

df['high_points'] = df.net_points > 2
df
```

```python
# df의 high_points를 삭제
del df['high_points']
df
```

```python
# zeros, net_points 제거
del df['zeros']
del df['net_points']
df
```

```python
# 새로운 행으로 데이터 넣기
# 2018, "Henry", 4.0, 0.1, 2.1 
df.loc['six',:] = [2018, "Henry", 4.0, 0.1, 2.1]
df
```
