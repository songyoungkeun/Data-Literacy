### 열 추가

```python
# df2에 국어 컬럼 생성과 데이터값을 동일하게 80으로 지정
df2.loc[:,'국어'] = 80
df2
```

```python
# 행 추가
df2.loc[3,:] = ['동규', 90,80,70,60,50]
df2
```

```python
# 행간 복사

df2.loc[4,:] = df2.loc[2,:]
df2
```

```python
# 이름 컬럼 -> index
df2.set_index(
    "이름",
    inplace=True
)
df2
```

```python
# 서준의 체육점수를 80으로 수정
df2.loc["서준","체육"] = 80
df2
```

```python
# Ex : 서준의 음악과 체육 점수를 50으로 변경하기
df2.loc["서준",["음악","체육"]] = 50
df2
```

```python
# 서준의 음악 : 100, 체육 : 50으로 수정

df2.loc["서준",["음악","체육"]] = 100, 50
df2
```

```python
# index를 초기화 
df2.reset_index(inplace=True)
df2
```
