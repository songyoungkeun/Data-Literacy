### 저장하기

```python
data = {
    "name" : ['Jerry', 'Riah', 'Paul'],
    "algol" : ['A', 'A+', 'B'],
    "basic" : ['C', 'B', 'B+'],
    "c++" : ['B+', 'C', 'C+'],
}
df = pd.DataFrame(data)
df.set_index('name', inplace=True)
df
```

```python
# csv로 저장하기
df.to_csv("../Data/df_sample.csv")
```

```python
# json으로 저장하기
df.to_json("../Data/df_sample.json")
```

```python
# Excel로 저장하기
df.to_excel("../Data/df_sample.xlsx")
```
